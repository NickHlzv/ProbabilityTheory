```python
#Объявим MSE
def mse_(B1, y=y, X=X, n=10):
    return np.sum((B1*X-y)**2)/n

alpha = 1e-6
alpha
```

```python
B1 = 0.1
n = 10
for i in range(800):
    B1 -=alpha*(2/n)*np.sum((B1*X-y)*X)
    print(f'B1={B1}')
```

```python
B1 = 0.1
for i in range(1000):
    B1 -=alpha*(2/n)*np.sum((B1*X-y)*X)
    if i%100==0:
        print(f'Iteration: {i}, B1={B1}, mse={mse_(B1)}')
```

Ответ 5.8898 совпадает с ранее посчитанными коэфициентами, значит подсчет корректный

```python
def mse_intercept(B0, B1, y=y, X=X, n=10):
    return np.sum((B0+B1*X-y)**2)/n
alpha = 1e-5
alpha
```

```python
B1 = 0.1
B0 = 1
for i in range(3000000):
    B1 -=alpha*(2/n)*np.sum(((B0+B1*X)-y)*X)
    B0 -=alpha*(2/n)*np.sum((B0+B1*X)-y)
    if i%100000==0:
        print(f'Iteration: {i}, B0={B0}, B1={B1}, mse={mse_intercept(B0, B1)}')

```
